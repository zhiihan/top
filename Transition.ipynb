```python
from scipy.optimize import curve_fit
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Just finite N scaling

xvals = np.linspace(0, 0.25, num=50)

eps = 0.001
plt.figure(figsize=(16,9))
delta_nums = []
m_data = []
b_data = []



for delta in np.linspace(0.5, 1.5, num=45):
    if df_all[abs(df_all['Delta']-delta)<eps].Delta.any():
        df = df_all[abs(df_all['Delta']-delta)<eps]
        fit, cov = curve_fit(lambda x, m, b: m*x+b, df['1/N'], df['M^2'])
        m_data.append(fit[0])
        b_data.append(fit[1])
        delta_nums.append(delta)
        plt.scatter(df['1/N'], df['M^2'])
    else: 
        print(f'No delta within {eps}, for delta = {delta}')
    
for m, b, delta in zip(m_data, b_data, delta_nums):
    plt.plot(xvals, m*xvals + b, label=f'$\Delta$ = {round(delta, 3)}')
plt.xlabel('$1/N$')
plt.ylabel('$M^2$')
plt.title('QuSpin Finite system scaling')
plt.legend()
plt.savefig('finitescale.png')
```
